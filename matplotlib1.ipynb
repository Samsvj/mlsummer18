```python
import matplotlib.pyplot as plt

plt.plot([1,2,3],[4,5,6])
plt.show()
```

```python
x=[1,2,3]
y=[4,5,6]

x2=[1,2,3]
y2=[10,14,12]

plt.plot(x,y, label='first line')
plt.plot(x2,y2, label='second line')

plt.xlabel('plot Number')

plt.ylabel('important var')

plt.title('interesting graphs \n check this out')
#The legend() method adds the legend to the plot. In
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),  shadow=True, ncol=2)

plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,3,5,7,9]
y=[5,3,7,2,8]

x2=[2,4,6,8,10]
y2=[10,14,12,15,17]

plt.bar(x,y, label='example one',color='red')
plt.bar(x2,y2, label='example two', color='green')

plt.legend()
plt.xlabel('bar number')

plt.ylabel('bar height')
plt.title('epic graphs \n another line! whoa')
plt.show()
```
